```python
from __future__ import print_function # adds comptability with python 2
import numpy as np    # useful for many scientific computing in python
import pandas as pd   # primary data structure library  
```

```python
df_can = pd.read_excel('C:/Users/aravind/Desktop/Study/Python/data_visulations_with_python/dataset/Canada.xlsx', 
                        'Canada by Citizenship', startcol = 0 , index = False,
                        skiprows=range(20), 
                        skipfooter=2)
        
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)

df_can.head(3)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)

df_can
```

```python
# examine the types of column labels
all(isinstance(column, str) for column in df_can.columns)
# returned false - tested to understand of the column labels are of type string
```

```python
#change them all to string type
df_can.columns = list(map(str, df_can.columns))

# examine the types of column labels
all(isinstance(column, str) for column in df_can.columns)

df_can
```

```python
# Set the country name as index - useful for quickly looking up countries using .loc method.
df_can.set_index('Country', inplace=True)
```

```python
df_can.head(3)
```

```python
# add a total column
df_can['Total'] = df_can.sum(axis = 1)
```

```python
df_can.head()
```

```python
# dataframe has 38 columns instead of 37 columns we had before 
print ('data dimensions:', df_can.shape)
```

```python
# finally, let's create a list of years from 1980 - 2013
# this will come in handy when we start plotting the data
years = list(map(str, range(1980, 2014)))
```

```python
years
```

```python
# let's view the 2013 data
df_can['2013'].head()
```

```python
#By default, the `histrogram` method breaks up the dataset into 10 bins. 
#The figure below summarizes the bin ranges and the frequency distribution of immigration in 2013. 
# We can see that in 2013:

# np.histogram returns 2 values
count, bin_edges = np.histogram(df_can['2013'])

print(count) # frequency count
print(bin_edges) # bin ranges, default = 10 bins
```

```python
"""
scatter Plot :  displays values pertaining to two , variables against each other
dependent variable plotted against an independent variable in order to determine 
if any correlation between the two variables exists 

"""

import matplotlib as mpl
import matplotlib.pyplot as plt 

df_total = df_can

# use the sum method to get the total population per year 
df_tot = pd.DataFrame(df_total[years].sum(axis = 0))

#change the years to int data type (for regression if needed)
df_tot.index = map(int, df_tot.index)

#reset the index to put in back in as a column in the dataframe 
df_tot.reset_index(inplace = True)

df_tot.columns = ['years', 'total']
```

```python
df_tot.head(1)
```

```python
df_tot.plot(kind='scatter', x='years', y='total', figsize=(9, 6), color='darkblue')
 
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
 
plt.show()
```
