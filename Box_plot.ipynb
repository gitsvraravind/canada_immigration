```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)

df_can.head(3)
```

```python
#change them all to string type
df_can.columns = list(map(str, df_can.columns))

# examine the types of column labels
all(isinstance(column, str) for column in df_can.columns)

df_can
```

```python
# Set the country name as index - useful for quickly looking up countries using .loc method.
df_can.set_index('Country', inplace=True)
```

```python
"""
Box Plot  # statistically representing the distribution of the given data through five main dimensions 

1st dimensions is minimum - smallest number in the sorted data

2nd dimension is the first quartile which is 25% of the way through sorted data i.e quarter of the data points are less than 
this value 

3rd dimension is the median which is meadian of the sorted data 

4th dimesnion is the third quartile which is 75% of the way through the sorted data i.e is three quarters of the data points are less 
than this value.

Final dimension is the maximum which is the highest number in the sorted data. 

"""
import matplotlib as mpl 
import matplotlib.pyplot as plt   


df_japan = df_can.loc[('Japan'), years].transpose()
df_japan = df_japan.plot(kind='box' , figsize = (9,6))
plt.title('Box Plot of Japanese immigrants [1980-2013]')
plt.ylabel('Number of immigrants')
plt.show() 

```
