```python
# remove the columns not required for analysis 
top5 = df_can.drop(['Type', 'Coverage', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName'], axis = 1)
top5.head(1)

top5.drop(['AREA'], axis =1 , inplace = True)
top5.head(1)

# rename the column odName to Country 
top5.rename(columns ={'OdName' : 'Country'}, inplace = True)
top5.head(1)
```

```python
top5.to_excel('C:/Users/aravind/Desktop/Study/Python/data_visulations_with_python/dataset/canada_imm.xls', index = False)
```

```python
top5.head()
```

```python
top5['Total'] = top5.iloc[:,-34:-1].sum(axis = 1)
top5
```

```python
top5.sort_values(['Total'], ascending = False, axis = 0, inplace = True)

can_imm_top5 = top5.head()

can_imm_top5
```

```python
can_imm_top5 = can_imm_top5.set_index(['Country'])
```

```python
years = list(range(1980,2014))
years
```

```python
can_imm_top5 = can_imm_top5[years].transpose()
```

```python
can_imm_top5
```

```python
# Generating Area Plots

import matplotlib as mpl
import matplotlib.pyplot as plt 

can_imm_top5.plot(kind = 'area')

plt.title("Immigration trend of top 5 countries")
plt.ylabel("Number of immogrants")
plt.xlabel("Years")

plt.show()
```
